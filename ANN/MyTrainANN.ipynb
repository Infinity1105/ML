```python
import tensorflow as  tf

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
data=pd.read_csv('heart.csv')
data
```

```python
x=data.iloc[0:,0:13].values
y=data.iloc[0:,13:14].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow import keras
from keras.layers import Dense
from keras import Sequential
```

```python
model=Sequential([
    Dense(units=13, activation='relu',kernel_initializer="uniform", name='layer_0'),
    Dense(units=26,activation='relu',kernel_initializer="uniform",name='layer_1'),
    Dense(units=1,activation='linear',kernel_initializer="uniform",name='layer_3'),
])
```

```python
from keras.optimizers import Adam
from keras.losses import BinaryCrossentropy
```

```python
model.compile(optimizer=Adam(),loss=BinaryCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50)
```

```python
logits=model(X_test)
y_pred=tf.nn.sigmoid(logits)
y_pred=(y_pred>0.5)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] + cm[0][0] + cm[1][1])
print(accuracy*100)
```
